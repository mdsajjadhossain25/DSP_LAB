```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def unit_step(start, end, sampling_rate):
    x = np.linspace(start=start,stop=end,num=int((end-start)*sampling_rate) + 1)
    y = np.where(x < 0, 0, 1)
    
    plt.title(f'Unit Step: {start} to {end})')
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.stem(x,y)
    plt.grid()
    plt.show()
```

```python
unit_step(-5, 4, 3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
def cosine_wave():
    t = np.linspace(0,0.02 , 1000)
    x = 10 *  np.cos(2 * np.pi * 500 * t + np.pi/2)

    plt.figure(figsize=(20,12))
    plt.subplot(4,1,1)
    plt.plot(t, x)
    plt.xlabel('Time(s)')
    plt.ylabel('Amplitude')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.grid(True)
    plt.show()
cosine_wave()
```

```python
def sine_wave():
    t = np.linspace(0,0.02 , 1000)
    x = 10 *  np.sin(2 * np.pi * 500 * t + np.pi/2)

    plt.figure(figsize=(20,12))
    plt.subplot(4,1,1)
    plt.plot(t, x)
    plt.xlabel('Time(s)')
    plt.ylabel('Amplitude')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.grid(True)
    plt.show()
sine_wave()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def ramp(start, end, sampling_rate):
    x = np.linspace(start=start, stop=end, num=int((end - start) * sampling_rate) + 1)
    y = np.where(x < 0, 0, x)
    
    # Plot the ramp function
    plt.figure(figsize=(8, 4))
    plt.title(f'Ramp Signal: {start} to {end}')
    plt.axhline(0, color='r', linewidth=0.5)
    plt.axvline(0, color='r', linewidth=0.5)
    plt.stem(x, y, basefmt=" ")
    plt.xlabel("Time")
    plt.ylabel("Amplitude")
    plt.show()

# Test the function
ramp(-5, 4, 3)
```

```python
def Exponential():
    t = np.linspace(-5, 4, 50)
    x = np.exp(t)
    plt.figure(figsize=(12,4))
    plt.title(f'Exponential')
    plt.stem(t, x)
    plt.xlabel('Time(s)')
    plt.ylabel('Amplitude')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.grid(True, which='both')
    plt.show()
Exponential()
```
