```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#plot original signal
def plot_original():
    f = 500
    t = np.linspace(0, 0.02, 1000)
    x = 5* np.sin(2 * np.pi * f * t + np.pi/2)
    
    plt.figure(figsize=(20, 12))
    
    plt.subplot(4, 1, 1)
    plt.plot(t, x)
    plt.title('Original Signal 1 (500 Hz)')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)
plot_original()
```

```python
def plot_oversampled():
    f = 500
    t = np.linspace(0, 0.02, 1000)
    x = 5 * np.sin(2 * np.pi * f * t + np.pi/2)
    
    fs = 8000
    t_sampled = np.arange(0, 0.02, 1/fs)
    x_sampled = 5 * np.sin(2 * np.pi * f * t_sampled + np.pi/2)

    # Plot
    plt.figure(figsize=(20, 12))
    plt.subplot(4, 1, 2)
    plt.stem(t_sampled, x_sampled)
    plt.title('Over-sampled Signal (8000 Hz)')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()

plot_oversampled()
```

```python
def plot_undersampled():
    f = 500
    t = np.linspace(0, 0.02, 1000)
    x = 5 * np.sin(2 * np.pi * f * t + np.pi/2)
    
    fs = 400
    t_sampled = np.arange(0, 0.02, 1/fs)
    x_sampled = 5 * np.sin(2 * np.pi * f * t_sampled + np.pi/2)

    # Plot
    plt.figure(figsize=(20, 12))
    plt.subplot(4, 1, 2)
    plt.stem(t_sampled, x_sampled)
    plt.title('Under-sampled Signal (400 Hz)')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()

plot_undersampled()
```

```python
t = np.linspace(0, 1, 1000)

x1 = np.sin(2 * np.pi * 10 * t)
x2 = np.sin(2 * np.pi * 50 * t)

fs = 40

x1_sampled = x1[::int(len(x1) / fs)]
x2_sampled = x2[::int(len(x2) / fs)]

# Plot the original and under-sampled signals
plt.figure(figsize=(20, 12))

plt.subplot(4, 1, 1)
plt.plot(t, x1)
plt.title('Original Signal 1 (10 Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(4, 1, 2)
plt.stem(t[::int(len(t) / fs)], x1_sampled)
plt.title('Under-sampled Signal 1 (10 Hz)')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(4, 1, 3)
plt.plot(t, x2)
plt.title('Original Signal 2 (50 Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(4, 1, 4)
plt.stem(t[::int(len(t) / fs)], x2_sampled)
plt.title('Under-sampled Signal 2 (50 Hz)')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
def plot_Nyquist():
    f = 500
    t = np.linspace(0, 0.02, 1000)
    x = 5 * np.sin(2 * np.pi * f * t + np.pi/2)
    
    fs = 2 * f
    t_sampled = np.arange(0, 0.02, 1/fs)
    x_sampled = 5 * np.sin(2 * np.pi * f * t_sampled + np.pi/2)

    # Plot
    plt.figure(figsize=(20, 12))
    plt.subplot(4, 1, 2)
    plt.stem(t_sampled, x_sampled)
    plt.title('Over-sampled Signal (2 * Fmax Hz)')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()

plot_Nyquist()
```
