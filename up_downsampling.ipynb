```python
import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(0, 0.02, 1000)
x = 5 * np.sin(2 * np.pi * 500 * t + np.pi / 2)

# Discrete signal
fs = 3750
n = np.arange(0, 0.02, 1 / fs)
x_n = 5 * np.sin(2 * np.pi * 500 * n + np.pi / 2)

# Upsampling by 2
l = 2
up_sampled = np.zeros(l * len(x_n))
up_sampled[::2] = x_n

#down_sampling by 2
down_sampled = x_n[::2]


# Plotting
plt.figure(figsize=(16, 10))

# Plot original signal
plt.subplot(4, 1, 1)
plt.plot(t, x)
plt.title(f"Original Signal")
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.axhline(color='r')
plt.axvline(color='r')

# Plot upsampled signal
plt.subplot(4, 1, 2)
plt.stem(up_sampled)
plt.title(f"Upsampled Signal")
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.axhline(color='r')
plt.axvline(color='r')

# Plot downsampled signal
plt.subplot(4, 1, 3)
plt.stem(down_sampled)
plt.title(f"Down sampled Signal")
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.axhline(color='r')
plt.axvline(color='r')

#smoothing the up sampled signal
for i in range(1, len(up_sampled)-1, 2):
    up_sampled[i]  = up_sampled[i-1]+up_sampled[i+1] /2

# Plot upsampled signal smothed
plt.subplot(4, 1, 4)
plt.stem(up_sampled)
plt.title(f"Upsampled Signal")
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.axhline(color='r')
plt.axvline(color='r')

plt.tight_layout()
plt.show()
```
